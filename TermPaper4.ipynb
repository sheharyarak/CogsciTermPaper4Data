```python
def activity(e_arr, start, end):
    index = np.arange(len(e_arr))
    nn = end - start
    act =  np.random.rand(nn, nn) * 40
    noise = np.random.normal(0, 0.1, e_arr.shape) * 70
    e_arrc = e_arr
    e_arrc[start:end, start:end] = act
    e_arrc +=  noise
    fdf = pd.DataFrame(data=e_arrc, index=index, columns=index)
    return fdf
```

```python
freqs = np.array([125, 250, 500, 1000, 1500, 2500, 4000, 6000, 8000, 10000, 12500, 15000])
freqs
```

```python
dataframes = {}
cx, cy = 14, 17
for i in range(len(freqs)):
    dataframe = activity(electrodes, cx - i, cy + i)
    plt.figure()
    sns.heatmap(data=dataframe, cmap='Blues', vmin=-70, vmax=70, cbar_kws={'label': 'Activity (mV)'})
    plt.xlabel('Electrode X')
    plt.ylabel('Electrode Y')
    plt.title(f'Recorded activity ({freqs[i]}Hz ± {freqs[i] / 5}Hz)')
    plt.savefig(f'{freqs[i]}Hz.png')
    dataframes[freqs[i]] = dataframe
```

```python
epochs = np.arange(1, 15000)

def f(x):
    if x < 3286:
        return 0.9 * (np.log10(0.01 * (x + 200)) - (-0.0000001 * ((x - 2750) ** 2) + 0.8))
    else:
        return 0.212 * np.log10(x) * 0.9 + 0.025 * np.sin(x / 1800)
    
train = (0.9 * (0.23 * np.log10(epochs)) + 0.025 * np.sin(epochs / 2000))
test = np.array([f(x) for x in epochs])

acc = pd.DataFrame(data={
    'epochs':epochs,
    'train': train,
    'test': test
})
acc.head(), max(train), max(test), min(train), min(test)
```

```python
sns.lineplot(data=acc, x='epochs', y='train', label='train')
sns.lineplot(data=acc, x='epochs', y='test', label='test')
plt.ylabel('accuracy')
plt.title('accuracy vs epochs [Fig 2]')
plt.savefig('accuracy.png')
```
